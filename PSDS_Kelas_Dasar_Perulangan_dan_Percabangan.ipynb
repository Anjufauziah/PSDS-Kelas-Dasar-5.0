#Looping/ Perulangan

Python menyediakan dua statement untuk melakukan proses perulangan yaitu for dan while. Diantara kedua statement ini, secara umum for lebih banyak digunakan daripada while.

#For Loop

Perulangan ```for ``` digunakan untuk melakukan iterasi didalam ```sequence``` seperti ``` list``` , ``` tuple```, ```string```, dan ```array``` atau objek yang dapat diiterasi lainnya.
**For loop syntax**
```
for val in sequence:
  loop body
  ```
**For Loop Flowchart**

WhatsApp Image 2021-10-19 at 17.37.59.jpeg

**Example: Python for loop**


````python
#Program untuk menghitung jumlah semua angka dalam sebuah list
L1 = [2,4,5,7,9]
jumlah = 0

#Iterasi di dalam list
for i in L1:
  jumlah = jumlah + i

print(jumlah)
````

#The range() function

````python
#Program untuk melakukan iterasi di dalam list melalui indeks
minuman = ["teh", "boba", "kopi","susu"]
p = len(minuman)

#Iterasi di dalam list melalui indeks
for i in range(p):
  print(minuman[i])
````

#for loop with else

````python
#Contoh for loop menggunakan else
for i in range(p):
  print(minuman[i])
else:
  print("Elemen sudah habis")
````

Diberikan ```list_tagihan```. kamu diminta untuk menghitung jumlah seluruh tagihan!. Gunakan looping ```for``` untuk menyelesaikan tugas ini! Simpan hasilnya pada variabel ```list_tagihan```

list_tagihan = [5000, 75000, -150000,125000,300000,-50000,200000]

#While loop

While looppada Python digunakan untuk melakukan iterasi pada blok kode selama nilai dari ```test expression``` adalah ```true```.

**Syntax of while loop in Python

```
while test_expression
  body of while
```

**While Loop Flowchart**

WhatsApp Image 2021-10-19 at 17.37.59 (1).jpeg

**Example: Python while Loop**


````python
# Program untuk menjumlahkan semua
# bilangan asli hingga suku ke-n
# jum = 1+2+3+...+n

n = 10
jumlah = 0
i = 1

while i<=n:
  jumlah = jumlah+i
  i = i+1

print(jumlah)
````

#While Loop with else

sama dengan ```for``` loop kita juga dapat mengkombinasikan ```while``` loop dengan blok ```else```.

Bagian ```else``` akan dieksekusi jika nilai pada ```test expression``` pada ```while loop``` bernilai ```false```. Perhatikan contoh berikut:

````python
jum = 0
i = 1
while i<=n:
  jum = jum+i
  i = i+1
else:
  print(jum) #dieksekusi diluar looping
````

#Latihan 2

Diberikan list ```tagihan``` hitunglah ```total_tagihan``` menggunakan while loop.

```tagihan = [50000, 75000, 125000, 300000, 200000]```

````python
#tuliskan kodemu
tagihan = [50000, 75000, 125000, 300000, 200000]
jumlah_tagihan = len(tagihan)
total_tagihan = 0
i = 0

while i<jumlah_tagihan:
  total_tagihan = total_tagihan + tagihan[i]
  i = i+1
print('total tagihan :', total_tagihan)
````
